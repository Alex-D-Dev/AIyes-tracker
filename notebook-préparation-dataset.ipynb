```python
# remove crop on eyes and resize 
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
path = './images/images_capture_crop'
arr = os.listdir(path)
for i in range (0,len(arr)-1):
    print(i)
    image = cv2.imread(path+'/'+arr[i])
    image2 = image[2:image.shape[0]-2 , 2:image.shape[0]-2]

    width = 32
    height = 32

    dim = (width, height) 
    image3 = cv2.resize(image2, dim, interpolation = cv2.INTER_AREA)
    cv2.imwrite('./images/images_capture_resize/'+str(arr[i]), image3)
```

```python
# creating dataset from images and features

file = open('./eyes_data.csv','w')

title = ''
for x in range (0,32):
    for y in range (0,32):
        title = title+'pixel_'+str(x)+'_'+str(y)+','
title = title+'id,milieu_crop_x,milieu_crop_y,distance,milieu_oeil_x,milieu_oeil_y,taille_original,cote,prediction_x,prediction_y'
file.write(title+'\n')

path = './images/images_capture_resize'
arr = os.listdir(path)
ref = open('./images/images_capture_crop/ref-crop.txt','r')
for i in range (0,len(arr)): 
    line = str(ref.readline())
    print(line)

    to_find = '_midcrop-'
    number = line[13 :line.find(to_find)]
            
    predict = number[number.find('-')+1:]
    predict_x = predict[0:predict.find('-')]
    predict_y = predict[predict.find('-')+1:]
            
    line = line[line.find(to_find)+len(to_find):len(line)]

    to_find ='.'
    midcrop_x = line[0:line.find(to_find)]
    #print(midcrop_x)
    line = line[len(str(midcrop_x))+len(to_find):len(line)]

    to_find =','
    line = line[line.find(to_find)+len(to_find):len(line)]

    to_find ='.'
    midcrop_y = line[0:line.find(to_find)]
    #print(midcrop_y)

    to_find ='-dist-'
    line = line[line.find(to_find)+len(to_find):len(line)]

    to_find ='-mideye-'
    distance = line[0:line.find(to_find)]
    distance_format = distance+'00000'
    distance_format = distance[0:5]
    #print(distance_format)
    line = line[len(str(distance))+len(to_find):len(line)]

    to_find ='.'
    mideye_x = line[0:line.find(to_find)]
    #print(mideye_x)
    line = line[len(str(midcrop_x))+len(to_find):len(line)]

    to_find =','
    line = line[line.find(to_find)+len(to_find):len(line)]

    to_find ='.'
    mideye_y = line[0:line.find(to_find)]
    #print(mideye_y)

    to_find ='-shape-('
    line = line[line.find(to_find)+len(to_find):len(line)]

    to_find =','
    original_shape = line[0:line.find(to_find)]
    line = line[len(str(original_shape))+len(to_find):len(line)]
    original_shape = str(int(original_shape)-4)
    #print(original_shape)

    to_find = '-'
    side = line[line.find(to_find)+len(to_find) : len(line)]
    to_find = '.png'
    side = side[0:side.find(to_find)]
    #print(side)

    total_to_add = ','+number+'-'+side+','+midcrop_x+','+midcrop_y+','+distance_format+','+mideye_x+','+mideye_y+','+original_shape+','+side+','+predict_x+','+predict_y
    print(total_to_add)

    image = cv2.imread(path+'/'+str(arr[i]))
    #print(image.shape)
    image = image.flatten()
    image2 =[]
    for k in range(0,image.shape[0]):
        if k%3 == 0:
            image2.append(image[k])
    #print(len(image2))
    image2 = str(image2)
    image3 = image2[1:len(image2)-1]+total_to_add
    #print(image3)
    file.write(image3+'\n')
ref.close() 
file.close()
```
